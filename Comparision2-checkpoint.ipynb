```python
print("Enter the name of file1")
f1 = input()
print("Enter the name of file2")
f2 = input()
```

```python
import re
def printFeatures1():
    with open(f1) as file1:
        features1 = []
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)
                words1 = line.split() 
                features1.append(words1[0])
        print("\nFEATURES OF PRODUCT1")
        for x in range(len(features1)): 
            print(features1[x])
        print('\n')
        
```

```python
def printFeatures2():
    with open(f2) as file2:
        features2 = []
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)1
                words2 = line.split()
                features2.append(words2[0])
        print("\nFEATURES OF PRODUCT2")
        for x in range(len(features2)): 
            print(features2[x])
        print('\n')
```

```python
def commonFeatures():
    features1 = []
    features2 = []
    with open(f1) as file1,open(f2) as file2:
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)
                words1 = line.split() 
                features1.append(words1[0])
       
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)1
                words2 = line.split()
                features2.append(words2[0])
    
        features=[]
        print("\nCOMMON FEATURES AMONG THE TWO PRODUCTS")        
        for feature in features2:
            if feature in features1:
                features.append(feature)
                #print(feature)
        
        for x in range(len(features)): 
            print(features[x])
        print('\n')
```

```python
def uncommonFeatures():
    features1 = []
    features2 = []
    with open(f1) as file1,open(f2) as file2:
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)
                words1 = line.split() 
                features1.append(words1[0])
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                #print(line)1
                words2 = line.split()
                features2.append(words2[0])
        product1=[]
        for feature in features1:
            if feature not in features2:
                product1.append(feature)
                #print(feature)
        print("\nFEATURES IN PRODUCT1 BUT NOT PRODUCT2")
        for x in range(len(product1)): 
            print(product1[x])
        print('\n')

        product2=[]
        for feature in features2:
            if feature not in features1:
                product2.append(feature)
                #print(feature)
        print("\nFEATURES IN PRODUCT2 BUT NOT PRODUCT1")
        for x in range(len(product2)): 
            print(product2[x])
        print('\n')
        
        
```

```python
def displayCommonFeatures():
    #comparison of 2 products
    a = []
    b = []
    c = []
    features1 = []
    features2 = []
    features = []

    with open(f1) as file1,open(f2) as file2:
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words1 = line.split()
                features1.append(words1[0])
                
                
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words2 = line.split()
                features2.append(words2[0])
               
                
        for feature in features2:
            if feature in features1:
                features.append(feature)
                
    with open(f1) as file1,open(f2) as file2:
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words1 = line.split() 
                if words1[0] in features:
                    a.append(words1[1])
    
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words2 = line.split() 
                if words2[0] in features:
                   b.append(words2[1])
                
        for feature in features:
            c.append(feature)
        
    print('\nCOMPARISON OF THE COMMON FEATURES') 
    for i in range(len(a)):
        print(c[i] + '\nProduct1: ' + a[i] + '\nProduct2: ' +  b[i] + '\n\n')
    print('\n')
```

```python
def  displayUncommonFeatures():
    #uncommon
    a = []
    b = []
    c = []
    features1 = []
    features2 = []
    with open(f1) as file1,open(f2) as file2:
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words1 = line.split()
                features1.append(words1[0])
                
                
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words2 = line.split()
                features2.append(words2[0])
    
    with open(f1) as file1,open(f2) as file2:
        product1=[]
        for feature in features1:
            if feature not in features2:
                product1.append(feature)
                #print(feature)
       
        product2=[]
        for feature in features2:
            if feature not in features1:
                product2.append(feature)
                #print(feature)
        
        for line in sorted(file1):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words1 = line.split() 
                if words1[0] in product1:
                    a.append(words1[1])
                
        for line in sorted(file2):
            line=line.rstrip()
            if re.search('[a-z]+-',line):
                words2 = line.split() 
                if words2[0] in product2:
                    b.append(words2[1])
                
        print('\nFEATURES IN PRODUCT1 BUT NOT PRODUCT2')
        for i in range(len(a)):
            print(product1[i] + ": "+ a[i])
     
        print('\nFEATURES IN PRODUCT2 BUT NOT PRODUCT1')
        for i in range(len(b)):
            print(product2[i] + ": "+ b[i])
        print('\n')
        
```

```python
print('Choose the desired option')
print('1.Features of Product1\n'+ 
      '2.Features of Product2\n'+
      '3.Common Features\n'+
      '4.Uncommon Features\n'+
      '5.Show the common features\n'+
      '6.Show the uncommon features\n')
while(1):
    choice = int(input())
    if choice == 1:
        printFeatures1()
    elif choice == 2:
        printFeatures2()
    elif choice == 3:
        commonFeatures()
    elif choice == 4:
        uncommonFeatures()
    elif choice == 5:
        displayCommonFeatures()
    elif choice == 6:
        displayUncommonFeatures()
    else:
        break;
```
